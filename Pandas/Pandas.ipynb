# Outliers

```python
#   lower half      upper half
#   -----------    -----------
#  |          |   |          |
# 45 47 52 52 53 55 56 58 62 80
#       |     |  |     |
#      Q2     ---      Q3
#      25%    Q2      75%
#             50%
# TQR=Q3-Q1=6
# Low val= Q1-1.5*IQR
# high val= Q3+1.5*IQR
#
# IQR = Inter Qurfile range
```

```python
def find_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-(1.5*IQR)
    high_val=q3+(1.5*IQR)
    
    print("Low Value: ",low_val,"High Value :",high_val)
    
    ds=ds.loc[(ds[col]<low_val) | (ds[col]>high_val)]
    return ds
```

```python
def remove_outliers(ds,col):
    q1=ds[col].quantile(0.25)
    q3=ds[col].quantile(0.75)
    IQR=q3-q1
    low_val=q1-(1.5*IQR)
    high_val=q3+(1.5*IQR)
    
    print("Low Value: ",low_val,"High Value :",high_val)
    
    ds=ds.loc[(ds[col]>=low_val) & (ds[col]<=high_val)]
    return ds
```

```python
import pandas as pd
fd=pd.read_csv("auto-mpg.csv")
find_outliers(fd,"mpg")
```

```python
find_outliers(fd,"acceleration")
```

```python
import pandas as pd
fd=pd.read_csv("auto-mpg.csv")
remove_outliers(fd,"mpg")
```

```python
df=pd.DataFrame({
    "Department":["Sales","Sales","HR","HR","IT","IT"],
    "Salary":[50000,60000,40000,45000,70000,75000],
    "Bonus":[5000,6000,2000,2500,8000,8500]
})
```

```python
df
```

```python
result=df.groupby("Department").agg("mean")
result
```

```python
result=df.groupby("Department").agg(["mean","max","min"])
result
```

```python
result=df.groupby("Department").agg({"Salary":["mean","max"],"Bonus":"sum"})
result
```
